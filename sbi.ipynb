```python
# Linear Regression 
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd


x = df[['days']]      
y = df['AVERAGE']   

model = LinearRegression()
model.fit(x, y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_

print(f"y = {slope}x + {intercept}")
```

```python
df['predicted'] = model.predict(x)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.plot(x, df['predicted'])
plt.xlabel('Days')
plt.ylabel('Average')
plt.title('Linear Regression')
plt.show()
```

```python
from sklearn.metrics import r2_score

print("R2 Score:", r2_score(y, df['predicted']))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_true = df['AVERAGE']
y_pred = df['predicted']

mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R2 Score:", r2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

# prepare features and target (use existing df)
X_nl = df[['days']]
y_nl = df['AVERAGE']

# train / test split
X_train, X_test, y_train, y_test = train_test_split(
    X_nl, y_nl, test_size=0.2, random_state=42, shuffle=True
)

# define a few non-linear candidate models / pipelines
models = {
    'RandomForest': RandomForestRegressor(n_estimators=200, random_state=42),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=200, random_state=42)
}

# fit, predict and compute metrics
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results.append({
        'model': name,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'R2': r2
    })

results_df = pd.DataFrame(results).sort_values('RMSE').reset_index(drop=True)
print(results_df)
```
