How to fill values 

    -> forwardfill => sequential repeated values
    
    -> backwardfill => sequential repeated values
    
    -> fillna => if we know what to fill
    
    -> replace => replacing the null values wi=th someother value
    
    -> mean => replacing with mean if any extreme values then use median
    
    -> median => filling null value with middle value
    
    -> mode => using relation formula r most repeated one
    
    -> random => In set of values choosing one value
    
    -> imputer methods> sklearn 
    
    -> min -> minimum 
    
    -> max -> maximum value in the set of data

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer, KNNImputer,IterativeImputer
from sklearn.experimental import enable_iterative_imputer

# -----------------------------
# Create small sample dataset
# -----------------------------
data = {
    "Age": [25, 30, np.nan, 40, 35],
    "Salary": [50000, np.nan, 60000, 80000, np.nan],
    "Experience": [1, 3, 5, np.nan, 4]
}

df = pd.DataFrame(data)

print("Original Data")
print(df)


# -----------------------------
# 1. Simple Imputer (Mean)
# -----------------------------
simple_imputer = SimpleImputer(strategy="mean")
df_simple = pd.DataFrame(simple_imputer.fit_transform(df))

print("\nSimple Imputer Result")
print(df_simple)


# -----------------------------
# 2. KNN Imputer
# -----------------------------
knn_imputer = KNNImputer(n_neighbors=2)
df_knn = pd.DataFrame(knn_imputer.fit_transform(df))

print("\nKNN Imputer Result")
print(df_knn)


# -----------------------------
# 3. Iterative Imputer
# -----------------------------
iter_imputer = IterativeImputer(random_state=0)
df_iter = pd.DataFrame(iter_imputer.fit_transform(df))

print("\nIterative Imputer Result")
print(df_iter)
```
