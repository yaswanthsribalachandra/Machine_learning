```python
# Bivariate: Numerical - categorical
```

```python
hr_sal = df[df["dept"] == "HR"]["salary"]
it_sal = df[df["dept"] == "IT"]["salary"]
fin_sal = df[df["dept"] == "Finance"]["salary"]
```

```python
# Boxplot:

plt.boxplot([hr_sal, it_sal, fin_sal], labels = ["Hr", "IT", "Finance"])
plt.show()
```

```python
# pie chart:

salary_by_dept = df.groupby("dept")["salary"].sum()
salary_by_dept
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct = "%1.2f", shadow = True, explode = [0.1, 0, 0.1])
plt.axis("equal")
plt.show()
```

```python
# bar plot

hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)

plt.bar(["HR", "IT", "Finance"], [hr_mean, it_mean, fin_mean], color = ["blue", "orange", "green"])
plt.ylabel("Average Salary")
plt.title("Average Salary by Department")
plt.show()  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

datasets = ["tips","titanic","iris","penguins","diamonds","flights"]

for name in datasets:
    
    print("Dataset:", name)
    
    df = sns.load_dataset(name)
    print(df.head())
    
    # numerical columns
    num_cols = df.select_dtypes(include=['int64','float64']).columns
    
    # categorical columns
    cat_cols = df.select_dtypes(include=['object','category']).columns
    
    
    # Histogram for numerical columns
    for col in num_cols:
        sns.histplot(df[col], kde=True)
        plt.title(f"{name} - Histogram of {col}")
        plt.show()
    
    
    # Countplot for categorical columns
    for col in cat_cols:
        sns.countplot(x=df[col])
        plt.title(f"{name} - Countplot of {col}")
        plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

datasets = ["tips","titanic","iris","penguins","diamonds","flights"]

for name in datasets:
    
    print("Dataset:", name)
    
    df = sns.load_dataset(name)
    print(df.head())
    
    # numerical columns
    num_cols = df.select_dtypes(include=['int64','float64']).columns
    
    # categorical columns
    cat_cols = df.select_dtypes(include=['object','category']).columns
    
    
    # Histogram for numerical columns
    for col in num_cols:
        sns.histplot(df[col], kde=True)
        plt.title(f"{name} - Histogram of {col}")
        plt.show()
    
    
    # Countplot for categorical columns
    for col in cat_cols:
        sns.countplot(x=df[col])
        plt.title(f"{name} - Countplot of {col}")
        plt.show()
```
