Univariate analysis is the process of analyzing a single variable independently to understand its distribution, central tendency, spread, and underlying pattern.
It is usually the first step in Exploratory Data Analysis (EDA).




“Univariate analysis is the study of a single variable to understand its distribution, central tendency, and spread.
For numerical data, we use plots like histograms, KDE, boxplots, and ECDF to analyze distribution and detect outliers.
For categorical data, count plots help us understand frequency and class imbalance.
This step is crucial in EDA because it helps in identifying patterns, anomalies, and preparing data for modeling.”

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Numerical column
data = df['total_bill']

# Categorical column
cat = df['day']
```

1. Histogram
Shows frequency distribution using bins
Helps identify:
Skewness (left/right)
Spread
Modality (uni-modal, bi-modal)

“Histogram is useful to quickly understand how data is distributed and whether it follows a normal distribution.”

```python
sns.histplot(data, bins=10)
plt.show()
```

2. KDE Plot (Kernel Density Estimation)
Smooth representation of distribution
Removes binning limitation of histogram
“KDE gives a continuous probability density curve, making it easier to observe the underlying distribution.”

```python
sns.kdeplot(data)
plt.show()
```

3. Distribution Plot
Combination of Histogram + KDE
In modern seaborn → replaced by histplot(kde=True)
“Distplot was widely used earlier, but now replaced due to better control in newer functions.”

```python
sns.histplot(data, kde=True)
plt.show()
```

4. ECDF (Empirical Cumulative Distribution Function)
Shows cumulative probability
For any value x → gives proportion of data ≤ x

“ECDF is useful when we want percentile-based insights without assuming any distribution.”

```python
sns.ecdfplot(data)
plt.show()
```

5. Box Plot
Displays:
Median
Quartiles (Q1, Q3)
IQR (Interquartile Range)
Outliers
👉 Important formula:
IQR = Q3 - Q1
Outliers:
< Q1 - 1.5×IQR OR > Q3 + 1.5×IQR
👉 statement:
“Boxplot is highly effective for detecting outliers and understanding data spread without distribution assumptions.”

```python
sns.boxplot(x=data)
plt.show()
```

```python
sns.boxplot(y=data)
plt.show()
```

6. Violin Plot
Combines boxplot + KDE
Shows both:
Distribution shape
Statistical summary
“Useful when we want both density and quartile information together.”

```python
sns.violinplot(x=data)
plt.show()
```

7. Boxen Plot
Advanced version of boxplot
Better for large datasets
Shows more detailed quantiles
“Boxen plots provide more granular insight into distribution tails compared to traditional boxplots.”

```python
sns.boxenplot(x=data)
plt.show()
```

Qualitative Data (Categorical Variables)
Count Plot
Displays frequency of each category
Example:
Gender distribution
Product categories

“Countplot helps in identifying class imbalance, which is critical for classification problems.”

```python
sns.countplot(x=cat)
plt.show()
```

Bivariate analysis studies the relationship between two variables. Scatter plots and regression plots help understand correlation between numerical variables. Line plots are used for trends, while strip and swarm plots visualize distributions across categories. Advanced plots like relplot and catplot allow multi-dimensional analysis. Residual plots help evaluate regression model performance

1.Scatter Plot
Shows relationship between two numerical variables
Helps detect:
Positive/negative correlation
Outliers
“Scatter plot is the most basic way to visualize correlation between two continuous variables.”

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

2.line plot
Shows trends over ordered data (time, sequence)
Useful for:
Time series analysis
“Line plots are useful when data has an inherent order, especially time-based.”

```python
sns.lineplot(x='size', y='total_bill', data=df)
plt.show()
```

3.strip plot
Shows distribution of data points
Adds jitter to avoid overlap
Insight:
“Useful for small datasets to visualize individual observations.”

```python
sns.stripplot(x='day', y='total_bill', data=df,jitter=0.1)
plt.show()

#jittering the x-axis values
#x_new = x_original + random(-jitter, +jitter)
#Strip plot uses random jitter to spread points, whereas swarm plot uses an algorithm to avoid overlap completely."
```

4.Swarn plot

Similar to strip plot but:
No overlapping points
Better visualization of distribution
“Swarm plots provide a clearer distribution without overlapping points.”
it is idea is like beeswarm

```python
sns.swarmplot(x='day', y='total_bill', data=df)
plt.show()
```

5. Rel Plot (Relational Plot)
High-level function
Can create:
Scatter plot
Line plot
Supports multiple variables (hue, size)

“Relplot is a figure-level function for visualizing relationships with additional dimensions.”

```python
sns.relplot(x='total_bill', y='tip', hue='sex', data=df)
```

6. Cat Plot (Categorical Plot)
Used for categorical relationships
Supports:
box, violin, bar, strip, etc.
“Catplot is flexible for categorical comparisons with different plot types.”

```python
sns.catplot(x='day', y='total_bill', kind='box', data=df)
```

7. LM Plot (Linear Model Plot)
Scatter plot + regression line
Shows:
Trend
Confidence interval
“LM plot helps visualize linear relationships along with regression fitting.”

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```

8. Reg Plot
Similar to lmplot but:
Works on axes level
👉 Key difference:
lmplot → figure-level
regplot → axes-level

```python
sns.regplot(x='total_bill', y='tip', data=df)
plt.show()
```

9. Residual Plot
Helps check:
Model fit
Linearity
“Residual plots are used to validate regression assumptions and detect patterns in errors.”

```python
sns.residplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")


# correlation for numeric columns only
num_corr = df.select_dtypes(include='number').corr()
print("Numeric columns correlation:")
print(num_corr)

# If you want to include categorical columns, convert categories to integer codes first
df_encoded = df.copy()
for col in df.select_dtypes(include='category').columns:
    df_encoded[col] = df_encoded[col].cat.codes

print("\nCorrelation including categorical columns (encoded):")
print(df_encoded.corr())

# Visualize numeric correlation matrix
sns.heatmap(num_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation matrix (numeric columns)')
plt.show()
# This will display the correlation matrix for the DataFrame
```
